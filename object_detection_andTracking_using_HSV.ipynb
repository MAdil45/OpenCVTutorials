```python
import cv2 as cv
import numpy as np
```

```python
## here we will be using HSV - (Hue Saturation, Value) for detection
## for better understanding : https://learn.leighcotnoir.com/artspeak/elements-color/hue-value-saturation/ 
# lets start wroking on it

while True:
    frame = cv.imread("./opencv_data/data/smarties.png")
    ## lets convert the image into hsv
    hsv = cv.cvtColor(frame, cv.COLOR_BGR2HSV)
    
    ## lets define a lower bound of a blue color
    l_b = np.array([110,50, 50])
    
    #lets define a upper bound of a color
    u_b = np.array([130,255,255])
    
    ## lets get a threshold to only get the blue color by using cv.inRange()
    mask = cv.inRange(hsv, l_b, u_b)
    res = cv.bitwise_and(frame, frame, mask=mask)
    cv.imshow('Frame', frame)
    cv.imshow('Mask', mask)
    cv.imshow('Res', res)

    key = cv.waitKey(1)
    if key == 27:
        break
cv.destroyAllWindows()
```

```python
## lets create a trackbar for better understanding and selection of colors
## re-use the above code with few modifications
## here we will be using HSV - (Hue Saturation, Value) for detection
## for better understanding : https://learn.leighcotnoir.com/artspeak/elements-color/hue-value-saturation/ 
# lets start wroking on it
## first lets define a oncallback function for trackbars
def nothing(x):
    pass

# lets define a window
cv.namedWindow("Tracking")
cv.createTrackbar("LowerHue", "Tracking", 0, 255, nothing)
cv.createTrackbar("LowerSaturation", "Tracking", 0, 255, nothing)
cv.createTrackbar("LowerValue", "Tracking", 0, 255, nothing)
cv.createTrackbar("UpperHue", "Tracking", 255, 255, nothing)
cv.createTrackbar("UpperSaturation", "Tracking", 255, 255, nothing)
cv.createTrackbar("UpperValue", "Tracking", 255, 255, nothing)

while True:
    
    frame = cv.imread("./opencv_data/data/smarties.png")
    
    ## lets convert the image into hsv
    hsv = cv.cvtColor(frame, cv.COLOR_BGR2HSV)
    
    ## lets use the trackbars to get the lower and upper values
    l_h = cv.getTrackbarPos("LowerHue", "Tracking")
    l_s = cv.getTrackbarPos("LowerSaturation", "Tracking")
    l_v = cv.getTrackbarPos("LowerValue", "Tracking")
    u_h = cv.getTrackbarPos("UpperHue", "Tracking")
    u_s = cv.getTrackbarPos("UpperSaturation", "Tracking")
    u_v = cv.getTrackbarPos("UpperValue", "Tracking")
    ## lets define a lower bound of a blue color
    l_b = np.array([l_h,l_s, l_v])
    
    #lets define a upper bound of a color
    u_b = np.array([u_h,u_s,u_v])
    
    ## lets get a threshold to only get the blue color by using cv.inRange()
    mask = cv.inRange(hsv, l_b, u_b)
    
    res = cv.bitwise_and(frame, frame, mask=mask)
    
    cv.imshow('Frame', frame)
    cv.imshow('Mask', mask)
    cv.imshow('Res', res)

    key = cv.waitKey(1)
    if key == 27:
        break
cv.destroyAllWindows()
```
