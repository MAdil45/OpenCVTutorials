```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Here, we will be using to the appplications of matplotlib library
## it is generally used for 2D ploting and charting
## lets first read an image
img = cv.imread("./Lena_Copy.png", -1)

## lets display it via using both cv and matplotlib
cv.imshow("IMAGE", img)

## before displaying with plt, lets convert it into RGB, as cv read in BGR
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
## when we run the aboe code, we will see the x and y coordinates as an index
## to hid them, lets use plt.xsicks and plt.yticks
## reusing the above code
img = cv.imread("./Lena_Copy.png", -1)

## lets display it via using both cv and matplotlib
cv.imshow("IMAGE", img)

## before displaying with plt, lets convert it into RGB, as cv read in BGR
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img)
plt.xticks([]), plt.yticks([])
plt.show()

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
## in image thresholding, we used six different images in six different windows. 
## what if we want to display them in a single window.
## we can either create a big size np array and paste onto it, or simply we can use matplotlib.
# lets see how to use matplotlib.
img = cv.imread("gradient.png")

## lets use cv.THRESH_BINARY_INV - (It will reverse the "0" to "1" and vise versa.)
_, thr1 = cv.threshold(img, 50, 255, cv.THRESH_BINARY)
_, thr2 = cv.threshold(img, 200, 255, cv.THRESH_BINARY_INV)
_, thr3 = cv.threshold(img, 127, 255, cv.THRESH_TRUNC)
_, thr4 = cv.threshold(img, 127, 255, cv.THRESH_TOZERO)
_, thr5 = cv.threshold(img, 127, 255, cv.THRESH_TOZERO_INV)

## to display all of them in a single window, lets create a list of titles
titles = ['Original', 'BINARY', 'BINARY_INV', 'THRESH_TRUNC', 'THRESH_TOZERO', 'THRESH_TOZERO_INV']

## now lets create a list of images
images = [img, thr1, thr2, thr3, thr4, thr5]

## lets use a simple for loop to show all the images
for i in range(6):
    plt.subplot(2, 3, i+1), 
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    ## if you want to hid x and y labels
    plt.xticks([]), plt.yticks([])
plt.show()
                       
```
